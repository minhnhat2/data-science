```python
#pyplot API
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color="cyan",label="sin(x)",linestyle="dashed");  ##label : đặt tên cho kí hiệu
plt.plot(x,np.cos(x),color="yellow",label="cos(x)",linestyle="dotted");##
plt.title("Bảng biểu đồ") #đặt tên biểu đồ
plt.xlabel("Biến X")     #đặt tên cột dọc 
plt.ylabel("F(x)")     #đặt tên cột ngang
plt.legend()   # xuất ra ô tên kí hiệu
plt.grid() #Tạo lưới 
```
