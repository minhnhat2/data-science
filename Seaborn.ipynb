```python
# FacetGrid
# FacetGrid có thể được vẽ với tối đa ba chiều: hàng, cột và sắc độ.
# Cách tiếp cận chính để trực quan hóa dữ liệu trên lưới này là sử dụng phương thức FacetGrid.map().
```

```python
tips_df.head()
```

```python
# Khảo sát những người hút thuốc có đến nhiều hơn những người không hút không 
tips_fg = sns.FacetGrid(data=tips_df, row="smoker", col="time") # Vẽ biểu đồ thể hiện những người có và không hút thuốc và thời điểm 
tips_fg.map(sns.scatterplot,  'total_bill', 'tip'); # trục hoành : total_bill , trục tung : tip
 # Qua biểu đồ ta có thể thấy , những người không hút thuốc ít đến hơn 
 # Tổng bill càng cao , tiền tip càng cao , người ăn tối nhiều hơn người ăn trưa 
```

```python
# Trực quan hóa đơn hàng của nam và nữ 
new_fg = sns.FacetGrid(data=tips_df, col="sex")
new_fg.map(sns.scatterplot, 'total_bill', 'tip')
 # Nữ thường ít tới hơn và số tiền tip ít hơn nam 
```

```python
# Thay đổi vị trí 
new_fg = sns.FacetGrid(data=tips_df, col="sex",
                       col_order=["Female", "Male"])
new_fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
new_fg = sns.FacetGrid(data=tips_df, col="sex",
                       hue="smoker",
                       col_order=["Female", "Male"])
new_fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
# Thêm chú thích 
new_fg = sns.FacetGrid(data=tips_df, col="sex",
                       hue="smoker",
                       col_order=["Female", "Male"])
new_fg.map(sns.scatterplot, 'total_bill', 'tip')
new_fg.add_legend();
```

```python
# Thay đổi màu của smoker 
new_fg = sns.FacetGrid(data=tips_df, col="sex",
                       hue="smoker",
                       col_order=["Female", "Male"],
                       palette='Set2',
                       legend_out=True)
new_fg.map(sns.scatterplot, 'total_bill', 'tip')
new_fg.add_legend();
```

```python
# Thay đổi chiều cao và chiều rộng và vòng tròn lớn hơn 
new_fg = sns.FacetGrid(data=tips_df, col="sex",
                       hue="smoker",
                       col_order=["Female", "Male"],
                       palette='Set2',
                       height=4, aspect=1.4,     # Chiều cao 4, chiều rộng 1.4
                       legend_out=True)

new_fg.map(sns.scatterplot, 'total_bill', 'tip',s=100 ,edgecolor='blue', alpha=0.7)   # Vòng tròn lớn 100 , viền màu xanh đậm, độ trong suốt 0.7

new_fg.add_legend();
```

```python
# Cách khác 
kws= dict(s=100, edgecolor='blue', alpha=.7)
new_fg = sns.FacetGrid(data=tips_df, col="sex",
                       hue="smoker",
                       col_order=["Female", "Male"],
                       palette='Set2',
                       height=4, aspect=1.4,
                       legend_out=True)

new_fg.map(sns.scatterplot, 'total_bill', 'tip', **kws)

new_fg.add_legend();
```

Project : Trực quan hóa dữ liệu dinh dưỡng trên 80 sản phẩm ngũ cốc
 file csv: https://raw.githubusercontent.com/CodexploreRepo/data-science/main/Code/A6_Seaborn/data/cereal.csv

*   source: https://www.kaggle.com/crawford/80-cereals



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CodexploreRepo/data-science/main/Code/A6_Seaborn/data/cereal.csv')
df.head()
```

```python
fields = ['shelf', 'weight', 'cups', 'rating'] ##Xóa những dữ liệu không liên quan 
cereal_df_new = df.drop(fields, axis=1)        #
cereal_df_new.head()
```

```python
cereal_corr = cereal_df_new.corr() # Lấy những dữ liệu số học 
cereal_corr
```

```python
## one_like có thể tạo ma trận booleans (True, False) có cùng hình dạng với dữ liệu của chúng ta
ones_corr = np.ones_like(cereal_corr, dtype=bool)
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape
```

```python
# np's triu: chỉ trả về ma trận tam giác trên
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=cereal_corr, mask=mask)
```

```python
# Khi loại bỏ góc ba trên, đây vẫn là hai tế bào trống trong ma trận của chúng tôi (calo & vitamin)
adjusted_mask = mask[1:, :-1]
adjusted_mask
```

```python
adjusted_cereal_corr = cereal_corr.iloc[1:, :-1]
```

```python
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask)
```

```python
# Thay đổi kích thước 
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask)
```

```python
# In các giá trị từng ô và làm tròn 2 chữ số thập phân 
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask,
            annot=True, fmt=".2f")  ### fmt".2f" để làm tròn 2 số thập phân 
```

```python
#  Thay đổi màu và tạo ngăn cách màu trắng 
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask,
            annot=True, annot_kws={"fontsize":13}, fmt=".2f", cmap="Blues",
            vmin=-1, vmax=1,
            linecolor='white', linewidths=0.5);
```

```python
fig, ax = plt.subplots(figsize=(10,8))

#Phương pháp đó sử dụng màu HUSL, vì vậy bạn cần màu sắc, độ bão hòa và độ sáng.
#Tôi đã sử dụng hsluv.org để chọn màu của biểu đồ này.
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)

sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask,
            annot=True, annot_kws={"fontsize":13}, fmt=".2f", cmap=cmap,
            vmin=-1, vmax=1,
            linecolor='white', linewidths=0.5);
```

```python
fig, ax = plt.subplots(figsize=(10,8))

cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)

sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask,
            annot=True, annot_kws={"fontsize":13}, fmt=".2f", cmap=cmap,   # Cho chữ bên trong ô to hơn annot_kws={"fontsize":13}
            vmin=-1, vmax=1,
            linecolor='white', linewidths=0.5);

yticks = [i.upper() for i in adjusted_cereal_corr.index]
xticks = [i.upper() for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks, rotation=0, fontsize=13)           # In Hoa và làm to chữ 
ax.set_xticklabels(xticks, rotation=90, fontsize=13)          # Làm cho chữ in hoa và di chuyển 90 độ 
title = 'CORRELATION MATRIX\nSAMPLED CEREALS COMPOSITION\n'   # Tạo tiêu đề
ax.set_title(title, loc='left', fontsize=18)                  # Đặt kích thước và vị trí tiêu đề
```
